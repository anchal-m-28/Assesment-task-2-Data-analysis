#### *Reading, cleaning, analysing and saving dataset*

```python
import pandas as pd

murders_by_country_df = pd.read_csv(
                          'Data/murders_by_country.csv',  
                          )

murders_by_country_df = murders_by_country_df.drop(columns=['Unnamed: 1'])

murders_by_country_df[['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']] = murders_by_country_df[['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']].astype(int)

murders_by_country_df['Total'] = murders_by_country_df[['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', 'Unnamed']].sum(axis=1)

murders_by_country_df = murders_by_country_df.sort_values('Total', ascending=False)

murders_by_country_df.to_csv('Data/murders_by_country_updated.csv')

murders_by_country_df.to_csv(
                             'Data/murders_by_country_updated.csv', 
                             index=False
                            )

murders_by_country_df
```

#### *Creating new .csv file with top 5 countries*

```python
murders_by_country_df = pd.read_csv(
                          'Data/murders_by_country_updated.csv', 
                        )

murders_by_country_df = murders_by_country_df.head(5)

murders_by_country_df.to_csv(
                             'Data/murders_by_country_top_5.csv', 
                             index=False
                            )
```

#### *Creating new .csv file with bottom 5 countries*

```python
murders_by_country_df = pd.read_csv(
                          'Data/murders_by_country_updated.csv', 
                        )

murders_by_country_df = murders_by_country_df.tail(5)

murders_by_country_df.to_csv(
                             'Data/murders_by_country_bottom_5.csv', 
                             index=False
                            )
```

#### *Creating Matplotlib chart for all countries*

```python

murders_by_country_df = pd.read_csv(
                          'Data/murders_by_country_updated.csv', 
                        )

murders_by_country_df.plot(
                           kind='bar',
                           x='country',
                           y='Total',
                           color='blue',
                           alpha=0.3,
                           title='The number of total murders in 187 countries between 1990 and 2016'
                          )
```

#### *Creating Matplotlib chart for top 5 countries*

```python
murders_by_country_top_5_df = pd.read_csv(
                          'Data/murders_by_country_top_5.csv', 
                        )

murders_by_country_top_5_df.plot(
                                 kind='bar',
                                 x='country',
                                 y='Total',
                                 color='blue',
                                 alpha=0.3,
                                 title='The top 5 countries with the most murders between 1990 and 2016'
                                )
```

#### *Creating Matplotlib chart for bottom 5 countries*

```python
murders_by_country_bottom_5_df = pd.read_csv(
                          'Data/murders_by_country_bottom_5.csv', 
                        )

murders_by_country_bottom_5_df.plot(
                                 kind='bar',
                                 x='country',
                                 y='Total',
                                 color='blue',
                                 alpha=0.3,
                                 title='The top 5 countries with the least murders between 1990 and 2016'
                                )
```
