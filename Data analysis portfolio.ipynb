#### *Specify Requirements*

The functional requirements for my program are listed below:
- **Data Loading**
  - Description: My program should be able to load .csv files and display an error message if the file is missing or in the wrong format.
  - Input: To this to happen the user should input a .csv file and ensure that it stored in the correct location.
  - Output: Once the user inputs this the program to be able to successfully work with the file without any issues.
- **Data Cleaning**
  - Description: My program should be able to filter out any unnecessary rows or colomns and only use those which are required.
  - Input: The user should input a .csv without any missing values and try to limit the amount of unnecessary rows. 
  - Output: The program should then be able to filter out the data and only use the rows or colomns required. 
- **Data Analysis**
  - Description: My program should allow for the mean and median to be calculated without any errors. 
  - Input: The user should input a file in the proper format (.csv).
  - Output: The program should then use the data in that file to find the mean and the median of each year and country. 
- **Data Visualisation**
  - Description: My program should visualise the data using Pandas dataframes as well as Matplotlib chart (most likely a bar graph).
  - Input: The user should enter the .csv file with the data they want visualised.
  - Output: The program should display the data in a Pandas dataframe and the appropriate Matplotlib chart.
- **Data Reporting**
  - Description: After all the processes above have been completed my program should store the final dataset as a .csv file. 
  - Input: The user should input the location they would like to store the final dataset. 
  - Output: The program should then take that location and store the final dataset accordingly. 

#### *Use cases*

The use cases for each functional requirement are listed below:
- **Data Loading**
  - Actor: User
  - Goal: To load a dataset into the system without any errors. 
  - Preconditions: User has a dataset file ready in the proper format (.csv).
  - Main Flow:
    - User places the dataset for reading into the correct folder.
    - System validates the file format without any error messages. 
    - System loads the dataset and displays the information in a dataframe.
  - Postconditions: Dataset is loaded and ready for cleaning.
- **Data Cleaning**
  - Actor: User
  - Goal: To clean the file and reduce the information to a smaller amount.
  - Preconditions: The dataset is loaded and ready for cleaning. 
  - Main Flow:
    - User indicates what information is required from the data set.
    - System goes through the dataframe and finds the required information. 
    - System displays the needed information in a dataframe.
  - Postconditions: More precise dataset is loaded and ready for analysis.
- **Data Analysis**
  - Actor: User
  - Goal: To use the smaller dataset to find the information we originally wanted to attain (mean, median, mode, etc).
  - Preconditions: The more precise dataset is loaded and ready for analysis.
  - Main Flow:
    - User tells the system what is required from the smaller dataset (eg. mean).
    - System accepts that and calculates the appropriate outcome. 
    - System displays the required information.
  - Postconditions: Dataset has been analysed and is ready for visualisation. 
- **Data Visualisation**
  - Actor: User
  - Goal: To use the information gained and display it in a proper way. 
  - Preconditions: Dataset has been analysed and is ready for visualisation.
  - Main Flow:
    - User selects which type of Matplotlib chart would be best for the visualisation of the data. 
    - System takes that and creates the appropriate type of chart along with a Pandas dataframe. 
    - System displays the Pandas dataframe and the appropriate Matplotlib chart type. 
  - Postconditions: Pandas dataframe and Matplotlib chart type are loaded and ready for reporting. 
- **Data Reporting**
  - Actor: User
  - Goal: To store the new information in a proper location. 
  - Preconditions: Pandas dataframe and Matplotlib chart type are loaded and ready for reporting.
  - Main Flow:
    - User indicates where they would like the information stored.
    - System stores the information in the proper location.
    - System lets the user know that the information has been stored successfully.
  - Postconditions: New information has been stored properly in the correct location.

**Data Privacy of Source**: I sourced my data from kaggle, a website which has thousands of free datasets. In my particular dataset the specific identities of any individual's aren't revealed as the source ensured that any personal data went through a proper de-identification process to reduce the risk of any personal data getting hacked or stolen. **Application Data Privacy**: As mentioned above, the dataset that I am using doesn't contain any identifiable personal data as it has all gone through a proper de-identification process to follow legal guidelines and ethical considerations. Taking into consideration that none of the data in my dataset is actually identifiable, once I finishing analysing and visualising my dataset, there aren't really any particular responsibilities I that have to handle if I were to push this application out to the general public as I'm not dealing with any identifiable personal data. 

**Cyber Security**: There are several features my application should have, once it is available online, in order to maintain cybersecurity. Some of the essential features include user authentication, password hashing and encryption. User authentication is a process implemented by applications to ensure that a person attempting to access certain information is actually who they claim to be. This can be done using passwords, PINs and two-factor authentication. Password hashing is a process that involves turning users' passwords into a short string of letters and/or numbers using an encryption algorithm. This is done as a security measure to ensure that passords are protected if a website is hacked. Encryption is a security measure in which information is converted to ciphertext, ensuring only certain authorized people (with a specific key) can decipher the code and access the original plaintext information. Once online, all these measures should be incorperated into my application in order to protect users and their personal and confidential information from security threats and hackers.
